Part 1

```python
instructions = []
with open("../input/day18.txt") as f:
    for line in f:
        l = line.strip().split()
        instructions.append((l[0], int(l[1])))
instructions
```

```python
x, y = 0, 0
area = 0
for d, l in instructions:
    if d == 'U':
        area += x * l
        y += l
    elif d == 'D':
        area -= x * l
        y -= l
    elif d == 'R':
        x += l
    elif d == 'L':
        x -= l
    else:
        raise ValueError(d)
abs(area) + sum([l for d, l in instructions]) // 2 + 1
```

Part 2

```python
instructions = []
with open("../input/day18.txt") as f:
    for line in f:
        instructions.append(line.strip().split()[2][2:-1])
instructions
```

```python
x, y = 0, 0
area = 0
dist = 0
for s in instructions:
    l = int(s[:-1], 16)
    dist += l
    if s[-1] == '3':
        area += x * l
        y += l
    elif s[-1] == '1':
        area -= x * l
        y -= l
    elif s[-1] == '0':
        x += l
    elif s[-1] == '2':
        x -= l
    else:
        raise ValueError(s)
abs(area) + dist // 2 + 1
```
